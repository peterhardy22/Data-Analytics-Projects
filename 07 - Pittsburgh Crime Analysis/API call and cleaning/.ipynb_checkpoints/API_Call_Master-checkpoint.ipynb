```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import date
```

```python
#Reading in the original csv file
csv_path= "Resources/crime_data.csv"
csv_df = pd.read_csv(csv_path)
```

```python
#Getting rid any empty rows/info
df = csv_df.dropna(how = 'any')
```

```python
#Splitting "ARRESTTIME' into just a "Date" column to pull from DarkSky API
time_df = df[['ARRESTTIME']]
time_df = time_df.reset_index()
time_df = time_df[['ARRESTTIME']]


time_df['Date'] = ''

i = 0
while i < len(time_df['ARRESTTIME']):
    time_df['Date'][i] = time_df['ARRESTTIME'][i].split('T')[0] 
    i += 1
```

```python
#Getting unique dates in order to retrieve Temperature and Weather Condition for each date
uni_date = time_df['Date'].unique()
unique_dates = list(uni_date)
```

```python
#Now narrowing down RANGE of data for the correct dates that we want to pull (2/12/16 to 12/31/18)
uni_date_df = pd.DataFrame({"Date": uni_date})
uni_date_df2 = uni_date_df.loc[(uni_date_df["Date"] >= "2016-01-01") & (uni_date_df["Date"] <= "2019-01-01")] 
uni_date_df3 = uni_date_df2.sort_values(["Date"])
uni_date_df3.head()
```

```python
#Making this DF into a list so that it is easier to work with
uni_date2 = uni_date_df3['Date']
unique_dates = list(uni_date2)
```

```python
#Specifying number of dates to pull from API
unique_dates_lookup = unique_dates[0:905]
```

```python
#using Emma's key for this pull
query1 = "https://api.darksky.net/forecast/eb7ebe07825cd001acae88909ef2b35d/40,-80,"
query2 = date
query3 = "T12:00:00"
list_1 = []


for ix, date in enumerate(unique_dates_lookup):
    print ("Working on {}/{}".format(ix+1, len(unique_dates_lookup)))
    url = query1+date+query3
    
    response = requests.get(query1+date+query3).json()
    list_1.append(response)
```

```python
#Combining Temperature and Weather Condition into one DF 
combined_df = uni_date_df3.copy()
combined_df["Temperature"]=""
combined_df["Weather Condition"]=""
combined_df=uni_date_df3.reset_index(drop=True)
combined_df.head()
```

```python
#need this for loop to drop missing info from API pull
missing_data_list= []
for counter in range(905):
    try:
        combined_df['Weather Condition'][counter]=list_1[counter]['currently']['summary']
    except KeyError:
        combined_df['Weather Condition'][counter]=''
        missing_data_list.append(counter)
    try:
        combined_df['Temperature'][counter]=list_1[counter]['currently']['temperature']
    except KeyError:
        combined_df['Temperature'][counter]=''
        missing_data_list.append(counter)
```
