```python
import pandas as pd
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
```

```python
url = 'https://www.artprice.com/artprice-reports/the-contemporary-art-market-report-2018/top-500-contemporary-artists/top-500-contemporary-artists-{}-to-{}/'
```

```python
urls_to_scrape = []
step = 50
urls_to_scrape.append(url.format(1, step))
for i in range(1, 10):
    urls_to_scrape.append(url.format((i*1 * step) + 1, (i+1)*step))
urls_to_scrape
list_of_urls = urls_to_scrape
list_of_urls
#tables = pd.read_html(urls_to_scrape[i])
#tables
```

```python
combined_df = pd.DataFrame()
for i in range(10):
    new_df = pd.read_html(list_of_urls[i])
    df = pd.DataFrame(new_df[0])
    # df[~df.isin(['NaN'])].any(axis=1)
    df = df.iloc[:-1]
    combined_df = combined_df.append(df)
    i +=1
combined_df
```

```python
combined_df[['Artist_Name', 'Dates Living']]=combined_df['Artist'].str.split('(', expand=True,n=2)
combined_cols = ["Rank", "Artist_Name", "Turnover", "Sold lots", "Highest price"]
combined_df_transformed = combined_df[combined_cols].copy()

combined_df_transformed = combined_df_transformed.rename(columns={"Rank": "Top_500_Rank",
                                                            "Sold lots": "Sold Lots",
                                                            "Highest price": "Highest Price"})


combined_df_transformed.head()
```

```python
combined_df_transformed.to_csv(r'Resources/top_500_artists.csv',index=False)
```
